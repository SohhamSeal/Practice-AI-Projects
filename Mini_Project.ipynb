## Make a neural network to predict on the “target” column of the dataset

###  Steps to complete:

1. Specify Architecture

2. Compile the model

3. Fit the model


```
import pandas as pd
from keras.utils import to_categorical
```

```
df=pd.read_csv("/content/heart.csv")
df.shape
```

```
df.info()
```

### **The columns and their description are as follows:**

1.	age
2.	sex(1 = male; 0 = female)
3.	chest pain type (4 values)
4.	resting blood pressure
5.	serum cholestoral in mg/dl
6.	fasting blood sugar > 120 mg/dl
7.	resting electrocardiographic results (values 0,1,2)
8.	maximum heart rate achieved
9.	exercise induced angina
10.	oldpeak = ST depression induced by exercise relative to rest
11.	the slope of the peak exercise ST segment
12.	number of major vessels (0-3) colored by flourosopy
13.	thal: 3 = normal; 6 = fixed defect; 7 = reversable defect


```
df.head()
```

```
df.isna().sum() #checking for missing values
```

```
target=to_categorical(df["target"])
target.shape
```

```
features=df.drop("target",axis=1)
features
```

After doing some research, I've found out that the columns that have a number rating format input, other than the 'sex column', indicate more prone to heart diseases as the value of the number increase, therefore, we keep everything intact, other than the 'sex' column, as we do not want any gender bias in our model.

```
features['sex']=features['sex'].replace([1,0],['Male','Female'])
features
```

```
features=pd.get_dummies(features,columns=["sex"]) #one-hot method encoding
features
```

# **Specifying Architechture:**
Here we have finite number of target values(0 and 1), therefore we create a binary classfication model

```
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
model=Sequential()
model.add(Dense(60,activation="relu",input_shape=(14,)))
model.add(Dense(56,activation="relu"))
model.add(Dense(2,activation="softmax"))
```

# **Compiling the model:**

```
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

# **Fitting the model:**

```
model.fit(features,target,epochs=72)
```
